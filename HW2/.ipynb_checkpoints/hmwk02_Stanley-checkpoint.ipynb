```python
%matplotlib inline
import multiprocessing as mp
mp.set_start_method('fork')
from matplotlib.pyplot import *
from numpy import *
```

```python
import warnings
warnings.simplefilter("ignore")   # Suppress overflow run time warnings

def julia_set(ax,bx,ay,by, Nx, Ny, kmax,c):
    
    # Generate points in complex planes D
    xe = linspace(ax,bx, Nx+1)
    ye = linspace(ay,by, Ny+1)
    dx = (bx-ax)/Nx
    dy = (by-ay)/Ny
    xc = xe[:-1] + dx/2            
    yc = ye[:-1] + dy/2            

    # Set of initial values Z0 : zk is a Nx x Ny matrix
    zk = xc + yc[:, None] * 1j
    
    # Constant needed for Julia fractal : g(z) = z^2 + c
    C = zeros_like(zk) + c
    
    # Divergence criteria
    rho = 2.0
    
    # Vectorize the computation of g(z); Use 
    escape_time = zeros_like(zk,dtype=int) + kmax
    for n in range(kmax):
        escaped = less(escape_time,kmax)
        if all(escaped):
            break
        I = logical_and(greater(abs(zk), rho),logical_not(escaped))
        escape_time[I] = n
        notI = not_equal(I,True)
        zk[notI] = zk[notI]**2 + C[notI]
    Iz = equal(escape_time,kmax)
    nz = count_nonzero(Iz)  # Number of zero values who never escaped
    return escape_time,nz
```

## Problem 2 : Fractal generation (parallel)

<hr style="border:solid 2px coral"></hr>

Generate the fractal from problem #1 using a parallel code.  

* Use a Pool method.  Report your timing results.  

* Include in your timing results any post-processing needed to create full $M$ matrix needed to generate the fractal image. 


```python
nquads = 4

dex = linspace(ax0,bx0,nquads+1)
dey = linspace(ay0,by0,nquads+1)

N = res//nquads
assert N*nquads == res, 'res must be divisible by nquads'
```

```python
%%time

def create_square(args):
    ax,bx,ay,by,N,kmax,c = args
    M,nz = julia_set(ax, bx, ay, by, N, N, kmax,c)    
    return M,nz

data = []
for i in range(nquads):
    for j in range(nquads):
        ax,bx = dex[i],dex[i+1]
        ay,by = dey[j],dey[j+1]
        data.append((ax,bx,ay,by,N,kmax,c))
        

with mp.Pool(processes=12) as pool:
    results_async=pool.map_async(create_square,data) 
    pool.close()
    pool.join()

results=results_async.get()
  
F = empty((res,res),dtype=int)    

nzt = 0
for i in range(nquads):
    for j in range(nquads):
        M,nz = results.pop(0)
        F[j*N:(j+1)*N,i*N:(i+1)*N] = M
        nzt += nz

print(f"Number of values that did not escape {nzt}")  
```

```python
%%time

dpi = 16        


imshow(F,vmin=0,vmax=kmax/5,origin='lower',cmap=cm.RdBu)
gca().axis('off')

imsave("Problem2_parallel.png",F,vmin=0,vmax=kmax/5,cmap=cm.RdBu,dpi=dpi,origin='lower')
```

## DISCUSSION


Certain factors must be considered when choosing a Pool method for fractal generation.

Firstly, determine whether a synchronous or an asynchronous pool method best suits the task. Asynchronous pool methods return the result as soon as a job is completed. However, an asynchronous pool method would wait for all tasks to be completed before returning results. In practice, Asynchronous methods are preferred over synchronous methods because they run jobs in a shorter time.

Second, one should consider whether or not to employ a callback function. When a specific task is finished, a callback function is invoked. 
This can be helpful for tasks that depend on the outcomes of earlier jobs because the callback can be used to transfer those outcomes to subsequent jobs.

Finally, depending on your particular Pool method, the target function may be limited to the primary data type or the number of arguments needed for successful compilation. If the fractal generation problem requires multiple-input arguments, it may be necessary to use a Pool method that suits this requirement.


In this exercise, the synchronous method (starmap) outperforms the asynchronous method (starmap_async).  However, this may change as the complexity of the problem increases.

Generate a high resolution image of the Mandelbrot set and describe the connection between the Mandelbrot set and the Julia sets we have been creating above. 

* Report your timing results. 

```python
def mandel_set(x0,y0,kmax): 
    z0=complex(x0,y0)
    z = z0
    for t in range(kmax):
        if abs(z) > 2.0:
            return t
        z = z * z + z0
    return kmax

MAX = 500

n=4096
xc=-0.5
yc=0
size=2
```

```python
%%time
array = zeros((n,n),dtype=np.uint8)
for col in range(n):
    for row in range(n):
        x0 = xc - (size / 2) + (size * col / n)
        y0 = yc - (size / 2) + (size * row / n)
        result=mandel_set(x0,y0, MAX)
        array[row, col] = result
```

```python
imshow(array,vmin=0,vmax=MAX/5,origin='lower',cmap=cm.gray)
gca().axis('off')
imsave("Problem5_serrial.png",array,vmin=0,vmax=MAX/5,cmap=cm.gray,dpi=dpi,origin='lower')
```
